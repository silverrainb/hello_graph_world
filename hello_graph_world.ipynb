```python
import networkx as nx
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
style.use('fivethirtyeight')
%matplotlib inline
```

## Load Data

This data contains travel website's product ids in Barcelona. The source and target in data refer to first and second purchased item with a time interval over a day in between. This data is acquired to gauge whether there is a connection between products purchase order with anticipation of making better recommendations to lead sales.

```python
data = pd.read_csv("data/bcn_gephi.csv")
```

```python
print(data.shape)
```

```python
# Subset for better visualization in networkx
data = data[:500]
```

```python
source = np.unique(data['Source'])
nodes = source.tolist()
```

```python
subset = data[['Source', 'Target']]
edges = [tuple(x) for x in subset.values]
```

Create an empty graph.

```python
graph = nx.Graph()
```

Add the nodes to graph.

```python
graph.add_nodes_from(nodes)
```

Add the edges to graph.

```python
graph.add_edges_from(edges)
```

Check the degree of each of the verticies to see connections.

```python
nx.degree(graph)
```

Visualize the graph.

```python
nx.draw(graph, with_labels=True)
#plt.show()
plt.savefig("graph.png", dpi=500)
```

We can see in the above graph that product id 1334 and 2509 are related to many purchases. We could interpret that when users purchase 1334, 2509 products, we can suggest the linked products to users to increase the likelihood of cross-selling. 

Furthermore, specifically for Barcelona landing page, we can display 1334, 2509 products as leading items on a landing page, and it is likely to increase the user acquisition & activation rate. We could also experiment by using these products as a proxy -- at the user acquisition stage, if users appear to be interested in those two items, then, chances of their purchase is higher.
